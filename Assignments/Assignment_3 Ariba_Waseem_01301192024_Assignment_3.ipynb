```python
# 2. Create a function to calculate sum of first N natural numbers.

def sum_natural_numbers(n):
    sum = 0
    for i in range(1, n+1):
        sum += i
    return sum
sum_natural_numbers(19037)
```

```python
# 3. Create a function to reverse a number

def reverse_number(n):
    reverse = 0
    while n > 0:
        reverse = reverse * 10 + n % 10
        n //= 10
    return reverse
reverse_number(123)
```

```python
# 4. Create a function to count digits in a number

def count_digit(n):
  count = 0
  for i in str(n):
    count += 1
  return count
count_digit(123)
```

```python
# 5. Create a function to check a palindrome number.

def palindrome_number(n):
  if n < 0:
    print (False)
  original = n
  reverse = 0
  while original > 0:
    reverse = (reverse * 10) + (original % 10)
    original = original // 10
  return reverse == n
print(palindrome_number(123))
print(palindrome_number(121))
```

```python
# 6. Create a function to generate fibonacci series.

def fibonacci_series(n):
  a,b = 0,1
  for i in range(1, n+1):
    print(a, end = " ")
    a,b = b,a+b
fibonacci_series(10)
```
